# **PHASE 4 — TRANSFORMER FOR DNA SEQUENCES**

Prepare tokenized DNA (Transformer input)

```python
import numpy as np
import pandas as pd

df = pd.read_csv("tcga_brca_sequence_dataset.csv")

token_map = {"A":0, "T":1, "G":2, "C":3}

def tokenize(seq):
    return [token_map[b] for b in seq]

X_tok = np.array([tokenize(s) for s in df["sequence"]])

df["label"] = df["type"].apply(lambda x: 0 if "Silent" in x else 1)
y = df["label"].values

print(X_tok.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tok, y, test_size=0.2, stratify=y, random_state=42
)
```

Positional Embedding Layer

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer

class PositionalEmbedding(Layer):
    def __init__(self, length, vocab_size, embed_dim):
        super().__init__()
        self.token_emb = tf.keras.layers.Embedding(vocab_size, embed_dim)
        self.pos_emb = tf.keras.layers.Embedding(length, embed_dim)

    def call(self, x):
        positions = tf.range(start=0, limit=tf.shape(x)[-1])
        return self.token_emb(x) + self.pos_emb(positions)
```

Transformer block

```python
from tensorflow.keras.layers import Dense, Dropout, LayerNormalization, MultiHeadAttention

class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)

    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))
```

Build Transformer model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, GlobalAveragePooling1D

SEQ_LEN = 500
VOCAB = 4
EMBED = 64

inputs = Input(shape=(SEQ_LEN,))
x = PositionalEmbedding(SEQ_LEN, VOCAB, EMBED)(inputs)

x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)

x = GlobalAveragePooling1D()(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.4)(x)

outputs = Dense(1, activation="sigmoid")(x)

transformer = Model(inputs, outputs)

transformer.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

transformer.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early = EarlyStopping(patience=5, restore_best_weights=True)

history_t = transformer.fit(
    X_train, y_train,
    epochs=50,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early],
    verbose=1
)
```

```python
loss_t, acc_t = transformer.evaluate(X_test, y_test)
print("Transformer Test Accuracy:", acc_t)
```

# **PHASE 5 — HYBRID CNN + TRANSFORMER MODEL (RESEARCH NOVELTY)**

```python
from sklearn.model_selection import train_test_split

X_train_h, X_test_h, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

CNN feature extractor block

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv1D, MaxPooling1D

def cnn_feature_extractor(x):
    x = Conv1D(64, 7, activation="relu", padding="same")(x)
    x = MaxPooling1D(2)(x)

    x = Conv1D(128, 5, activation="relu", padding="same")(x)
    x = MaxPooling1D(2)(x)

    x = Conv1D(256, 3, activation="relu", padding="same")(x)
    return x
```

Transformer block (reuse structure)

```python
from tensorflow.keras.layers import Dense, Dropout, LayerNormalization, MultiHeadAttention

class HybridTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)

    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))
```

Build Hybrid Model

```python
from tensorflow.keras.layers import Input, GlobalAveragePooling1D
from tensorflow.keras.models import Model

inputs = Input(shape=(500,4))

# CNN stage
x = cnn_feature_extractor(inputs)

# Now x has shape (~125 timesteps, 256 channels)

# Transformer stage
x = HybridTransformerBlock(256, num_heads=4, ff_dim=256)(x)
x = HybridTransformerBlock(256, num_heads=4, ff_dim=256)(x)

# Classification head
x = GlobalAveragePooling1D()(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.4)(x)

outputs = Dense(1, activation="sigmoid")(x)

hybrid = Model(inputs, outputs)

hybrid.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

hybrid.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early = EarlyStopping(patience=5, restore_best_weights=True)

history_h = hybrid.fit(
    X_train_h, y_train,
    epochs=40,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early],
    verbose=1
)
```

```python
driver_genes = {
    "TP53","BRCA1","BRCA2","PIK3CA","PTEN","KRAS","EGFR",
    "RB1","AKT1","BRAF","CDH1","GATA3","MAP3K1","NF1",
    "NOTCH1","ERBB2","ATM","CHEK2"
}

df = pd.read_csv("tcga_brca_sequence_dataset.csv")

df["label"] = df["gene"].apply(lambda g: 1 if g in driver_genes else 0)

print("Driver mutations:", df["label"].sum())
print("Passenger mutations:", len(df) - df["label"].sum())
```

```python
from sklearn.utils import resample

df_pos = df[df.label == 1]
df_neg = df[df.label == 0]

df_neg_down = resample(
    df_neg,
    replace=False,
    n_samples=len(df_pos),
    random_state=42
)

df_balanced = pd.concat([df_pos, df_neg_down])
df_balanced = df_balanced.sample(frac=1, random_state=42)

print("Balanced dataset size:", len(df_balanced))
```

```python
def inject_mutation(row):
    seq = list(row["sequence"])
    center = len(seq)//2
    seq[center] = row["alt"]
    return "".join(seq)

df_balanced["sequence_mut"] = df_balanced.apply(inject_mutation, axis=1)
```

```python
import numpy as np

mapping = {
    "A":[1,0,0,0],
    "T":[0,1,0,0],
    "G":[0,0,1,0],
    "C":[0,0,0,1]
}

def one_hot(seq):
    return np.array([mapping[b] for b in seq])

X = np.array([one_hot(s) for s in df_balanced["sequence_mut"]])
y = df_balanced["label"].values

print("Input shape:", X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model

# -------------------------------
# Improved CNN Feature Extractor (With Residual Stabilization)
# -------------------------------
def cnn_feature_extractor(x):

    # Block 1
    x1 = Conv1D(64, 7, padding="same")(x)
    x1 = BatchNormalization()(x1)
    x1 = Activation("relu")(x1)
    x1 = MaxPooling1D(2)(x1)

    # Block 2
    x2 = Conv1D(128, 5, padding="same")(x1)
    x2 = BatchNormalization()(x2)
    x2 = Activation("relu")(x2)
    x2 = MaxPooling1D(2)(x2)

    # Block 3
    x3 = Conv1D(256, 3, padding="same")(x2)
    x3 = BatchNormalization()(x3)
    x3 = Activation("relu")(x3)

    # Residual projection for stability
    shortcut = Conv1D(256, 1, padding="same")(x2)
    x3 = Add()([x3, shortcut])

    x3 = Dropout(0.3)(x3)

    return x3


# -------------------------------
# Positional Encoding Layer (Modified for Feature Tensors)
# -------------------------------
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, max_len, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.pos_emb = Embedding(input_dim=max_len, output_dim=embed_dim)

    def call(self, x):
        # x shape: (batch_size, seq_len, embed_dim)
        seq_len = tf.shape(x)[1]
        positions = tf.range(start=0, limit=seq_len, delta=1)
        pos_encoding = self.pos_emb(positions)
        return x + pos_encoding


# -------------------------------
# Improved Hybrid Transformer Block (Lighter + Stable)
# -------------------------------
class HybridTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()

        self.att = MultiHeadAttention(
            num_heads=num_heads,
            key_dim=embed_dim // num_heads,
            dropout=0.3
        )

        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dropout(0.3),
            Dense(embed_dim)
        ])

        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.norm2 = LayerNormalization(epsilon=1e-6)

        self.drop1 = Dropout(0.3)
        self.drop2 = Dropout(0.3)

    def call(self, x, training=False):
        attn_output = self.att(x, x)
        attn_output = self.drop1(attn_output, training=training)
        x = self.norm1(x + attn_output)

        ffn_output = self.ffn(x, training=training)
        ffn_output = self.drop2(ffn_output, training=training)

        return self.norm2(x + ffn_output)


# -------------------------------
# Model Definition (Same variable names)
# -------------------------------
inputs = Input(shape=(X.shape[1], 4))

# CNN stage
x = cnn_feature_extractor(inputs)

# Positional encoding AFTER CNN compression
# The sequence length after CNN processing is 500 / 2 / 2 = 125
x = PositionalEncoding(max_len=125, embed_dim=256)(x)

# Reduced Transformer depth (1 block instead of 2)
x = HybridTransformerBlock(embed_dim=256, num_heads=2, ff_dim=128)(x)

x = GlobalAveragePooling1D()(x)

x = Dense(128, activation="relu")(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)

outputs = Dense(1, activation="sigmoid")(x)

hybrid = Model(inputs, outputs)

hybrid.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

hybrid.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early = EarlyStopping(patience=6, restore_best_weights=True)

history = hybrid.fit(
    X_train, y_train,
    epochs=50,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early],
    verbose=1
)
```

```python
loss, acc = hybrid.evaluate(X_test, y_test)
print("Improved Hybrid Accuracy:", acc)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def compute_saliency(model, x):
    x = tf.convert_to_tensor(x[None, ...], dtype=tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(x)
        pred = model(x)

    grads = tape.gradient(pred, x)
    saliency = tf.reduce_max(tf.abs(grads), axis=-1).numpy()[0]

    return saliency
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, Input, GlobalAveragePooling1D, LayerNormalization, MultiHeadAttention, Layer

# --- Load original sequence dataset for silent/non-silent classification ---
df_original = pd.read_csv("tcga_brca_sequence_dataset.csv")
df_original["label"] = df_original["type"].apply(
    lambda x: 0 if "Silent" in x else 1
)

# Prepare one-hot encoded data for CNN and Hybrid models
mapping = {"A":[1,0,0,0], "T":[0,1,0,0], "G":[0,0,1,0], "C":[0,0,0,1]}
def one_hot(seq):
    return np.array([mapping[b] for b in seq])
X_onehot_original = np.array([one_hot(seq) for seq in df_original["sequence"]])

# Prepare tokenized data for Transformer model
token_map = {"A":0, "T":1, "G":2, "C":3}
def tokenize(seq):
    return [token_map[b] for b in seq]
X_tokenized_original = np.array([tokenize(s) for s in df_original["sequence"]])

y_original = df_original["label"].values

# ==================================================================
# Create initial train/test splits for each model (silent/non-silent)
# These match the data used for the original training of each model.
# ==================================================================

# CNN and Hybrid (first version) shared the same one-hot encoded data split
X_train_cnn_initial, _, y_train_cnn_initial, _ = train_test_split(
    X_onehot_original, y_original, test_size=0.2, random_state=42, stratify=y_original
)

# Transformer used tokenized data split
X_train_trans_initial, _, y_train_trans_initial, _ = train_test_split(
    X_tokenized_original, y_original, test_size=0.2, random_state=42, stratify=y_original
)

# ==================================================================
# Further split training data into training and validation sets for timing
# ==================================================================

# CNN / Hybrid data split for timing
X_tr_cnn, X_val_cnn, y_tr_cnn, y_val_cnn = train_test_split(
    X_train_cnn_initial, y_train_cnn_initial,
    test_size=0.2,
    stratify=y_train_cnn_initial,
    random_state=42
)

# Transformer data split for timing
X_tr_trans, X_val_trans, y_tr_trans, y_val_trans = train_test_split(
    X_train_trans_initial, y_train_trans_initial,
    test_size=0.2,
    stratify=y_train_trans_initial,
    random_state=42
)


# --- Model Re-creation Functions (from previous cells) ---

def build_cnn_model():
    model = Sequential([
        Conv1D(64, 7, activation="relu", input_shape=(500,4)),
        MaxPooling1D(2),
        Conv1D(128, 5, activation="relu"),
        MaxPooling1D(2),
        Conv1D(256, 3, activation="relu"),
        MaxPooling1D(2),
        Flatten(),
        Dense(256, activation="relu"),
        Dropout(0.5),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

class PositionalEmbedding(Layer):
    def __init__(self, length, vocab_size, embed_dim):
        super().__init__()
        self.token_emb = tf.keras.layers.Embedding(vocab_size, embed_dim)
        self.pos_emb = tf.keras.layers.Embedding(length, embed_dim)
    def call(self, x):
        positions = tf.range(start=0, limit=tf.shape(x)[-1])
        return self.token_emb(x) + self.pos_emb(positions)

class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)
    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))

def build_transformer_model():
    SEQ_LEN = 500
    VOCAB = 4
    EMBED = 64
    inputs = Input(shape=(SEQ_LEN,))
    x = PositionalEmbedding(SEQ_LEN, VOCAB, EMBED)(inputs)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    transformer_model = Model(inputs, outputs)
    transformer_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return transformer_model


def cnn_feature_extractor(x):
    x = Conv1D(64, 7, activation="relu", padding="same")(x)
    x = MaxPooling1D(2)(x)
    x = Conv1D(128, 5, activation="relu", padding="same")(x)
    x = MaxPooling1D(2)(x)
    x = Conv1D(256, 3, activation="relu", padding="same")(x)
    return x

class HybridTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)

    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))

def build_hybrid_model():
    inputs = Input(shape=(500,4))
    x = cnn_feature_extractor(inputs)
    x = HybridTransformerBlock(256, num_heads=4, ff_dim=256)(x)
    x = HybridTransformerBlock(256, num_heads=4, ff_dim=256)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    hybrid_model = Model(inputs, outputs)
    hybrid_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return hybrid_model


# =========
# CNN timing
# =========

cnn_model_for_timing = build_cnn_model()
cnn_time = train_with_time(
    cnn_model_for_timing,
    X_tr_cnn, y_tr_cnn,
    X_val_cnn, y_val_cnn
)
cnn_params = count_params(cnn_model_for_timing)

# =================
# Transformer timing
# =================

transformer_model_for_timing = build_transformer_model()
trans_time = train_with_time(
    transformer_model_for_timing,
    X_tr_trans, y_tr_trans,
    X_val_trans, y_val_trans
)
trans_params = count_params(transformer_model_for_timing)

# ============)
# Hybrid timing
# ============)

hybrid_model_for_timing = build_hybrid_model()
hyb_time = train_with_time(
    hybrid_model_for_timing,
    X_tr_cnn, y_tr_cnn,
    X_val_cnn, y_val_cnn
)
hyb_params = count_params(hybrid_model_for_timing)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Input, GlobalAveragePooling1D, Dense, Dropout, Layer, LayerNormalization, MultiHeadAttention, Conv1D, MaxPooling1D, BatchNormalization, Activation, Add, Embedding
from tensorflow.keras.models import Sequential, Model
from sklearn.utils import resample # needed for df_balanced recreation

# --- Recreate df_balanced to get access to raw sequence_mut for tokenization ---
# This ensures that X_train/X_test derived here are consistent with the original notebook flow
driver_genes = {
    "TP53","BRCA1","BRCA2","PIK3CA","PTEN","KRAS","EGFR",
    "RB1","AKT1","BRAF","CDH1","GATA3","MAP3K1","NF1",
    "NOTCH1","ERBB2","ATM","CHEK2"
}
df_full = pd.read_csv("tcga_brca_sequence_dataset.csv") # Original full dataset
df_full["label_driver"] = df_full["gene"].apply(lambda g: 1 if g in driver_genes else 0)

df_pos = df_full[df_full.label_driver == 1]
df_neg = df_full[df_full.label_driver == 0]
df_neg_down = resample(
    df_neg,
    replace=False,
    n_samples=len(df_pos),
    random_state=42
)
es = df_neg_down
df_balanced_recreated = pd.concat([df_pos, df_neg_down])
df_balanced_recreated = df_balanced_recreated.sample(frac=1, random_state=42)

# inject_mutation is defined in cell Hu9lbHHbxXxo, define it here for self-containment
def inject_mutation(row):
    seq = list(row["sequence"])
    center = len(seq)//2
    seq[center] = row["alt"]
    return "".join(seq)

df_balanced_recreated["sequence_mut"] = df_balanced_recreated.apply(inject_mutation, axis=1)

# --- Define one_hot function for X_balanced_oh_full, from cell BO5XyEjYxZa9 ---
mapping = {
    "A":[1,0,0,0],
    "T":[0,1,0,0],
    "G":[0,0,1,0],
    "C":[0,0,0,1]
}
def one_hot(seq):
    return np.array([mapping[b] for b in seq])

# --- Prepare full one-hot encoded and tokenized balanced datasets ---
X_balanced_oh_full = np.array([one_hot(s) for s in df_balanced_recreated["sequence_mut"]])
y_balanced_full_labels = df_balanced_recreated["label_driver"].values

token_map = {"A":0, "T":1, "G":2, "C":3}
def tokenize_sequence(seq):
    return [token_map[b] for b in seq]
X_tok_balanced_full = np.array([tokenize_sequence(s) for s in df_balanced_recreated["sequence_mut"]])

# --- Perform the train_test_split to get consistent splits for all models ---
# We need to ensure that the X_train/X_test passed to evaluate_model are from this split.
# X_oh_train_for_models is for CNN/Hybrid, X_tok_train_for_models is for Transformer
# y_train_for_models is the common label set
X_oh_train_for_models, X_oh_test_for_models, \
    X_tok_train_for_models, X_tok_test_for_models, \
    y_train_for_models, y_test_for_models = train_test_split(
    X_balanced_oh_full, X_tok_balanced_full, y_balanced_full_labels,
    test_size=0.2,
    stratify=y_balanced_full_labels,
    random_state=42
)


# --- Model Builder Functions (Copied from previous executed cells for self-containment) ---

def build_cnn(seq_len):
    model = Sequential([
        Conv1D(64, 7, activation="relu", input_shape=(seq_len, 4)),
        MaxPooling1D(2),
        Conv1D(128, 5, activation="relu"),
        MaxPooling1D(2),
        Conv1D(256, 3, activation="relu"),
        MaxPooling1D(2),
        Flatten(),
        Dense(256, activation="relu"),
        Dropout(0.5),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

class PositionalEmbedding(Layer):
    def __init__(self, length, vocab_size, embed_dim):
        super().__init__()
        self.token_emb = Embedding(vocab_size, embed_dim)
        self.pos_emb = Embedding(length, embed_dim)
    def call(self, x):
        positions = tf.range(start=0, limit=tf.shape(x)[-1])
        return self.token_emb(x) + self.pos_emb(positions)

class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)
    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))

def build_transformer(seq_len):
    VOCAB = 4
    EMBED = 64
    inputs = Input(shape=(seq_len,))
    x = PositionalEmbedding(seq_len, VOCAB, EMBED)(inputs)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    transformer_model = Model(inputs, outputs)
    transformer_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return transformer_model

def cnn_feature_extractor(x_input):
    x1 = Conv1D(64, 7, padding="same")(x_input)
    x1 = BatchNormalization()(x1)
    x1 = Activation("relu")(x1)
    x1 = MaxPooling1D(2)(x1)
    x2 = Conv1D(128, 5, padding="same")(x1)
    x2 = BatchNormalization()(x2)
    x2 = Activation("relu")(x2)
    x2 = MaxPooling1D(2)(x2)
    x3 = Conv1D(256, 3, padding="same")(x2)
    x3 = BatchNormalization()(x3)
    x3 = Activation("relu")(x3)
    shortcut = Conv1D(256, 1, padding="same")(x2)
    x3 = Add()([x3, shortcut])
    x3 = Dropout(0.3)(x3)
    return x3

# PositionalEncoding for hybrid is different (for features, not tokens)
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, max_len, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.pos_emb = Embedding(input_dim=max_len, output_dim=embed_dim)
    def call(self, x):
        seq_len = tf.shape(x)[1]
        positions = tf.range(start=0, limit=seq_len, delta=1)
        pos_encoding = self.pos_emb(positions)
        return x + pos_encoding

class HybridTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim // num_heads, dropout=0.3)
        self.ffn = tf.keras.Sequential([Dense(ff_dim, activation="relu"), Dropout(0.3), Dense(embed_dim)])
        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.norm2 = LayerNormalization(epsilon=1e-6)
        self.drop1 = Dropout(0.3)
        self.drop2 = Dropout(0.3)
    def call(self, x, training=False):
        attn_output = self.att(x, x)
        attn_output = self.drop1(attn_output, training=training)
        x = self.norm1(x + attn_output)
        ffn_output = self.ffn(x, training=training)
        ffn_output = self.drop2(ffn_output, training=training)
        return self.norm2(x + ffn_output)

def build_hybrid(seq_len):
    inputs = Input(shape=(seq_len, 4))
    x = cnn_feature_extractor(inputs)
    x = PositionalEncoding(max_len=125, embed_dim=256)(x)
    x = HybridTransformerBlock(embed_dim=256, num_heads=2, ff_dim=128)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    hybrid = Model(inputs, outputs)
    hybrid.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), loss="binary_crossentropy", metrics=["accuracy"])
    return hybrid

# --- End Model Builder Functions ---


seeds = [1, 2, 3, 4, 5]

model_builders = {
    "CNN": build_cnn,
    "Transformer": build_transformer,
    "Hybrid": build_hybrid
}

all_results = []

for model_name, build_fn in model_builders.items():

    print(f"\n========== {model_name} ==========")

    for seed in seeds:
        print(f"Running Seed {seed}")

        tf.keras.backend.clear_session()
        tf.keras.utils.set_random_seed(seed)

        # Prepare data based on model_name
        if model_name == "Transformer":
            X_train_data = X_tok_train_for_models
            X_test_data = X_tok_test_for_models
            input_seq_len = X_tok_train_for_models.shape[1]
        else: # CNN and Hybrid use one-hot encoded data
            X_train_data = X_oh_train_for_models
            X_test_data = X_oh_test_for_models
            input_seq_len = X_oh_train_for_models.shape[1]

        model = build_fn(input_seq_len) # Pass the correct seq_len

        # Create validation split from the current training data (X_train_data, y_train_for_models)
        X_train_fold, X_val_fold, y_train_fold, y_val_fold = train_test_split(
            X_train_data, y_train_for_models,
            test_size=0.1, # Using 10% for validation, consistent with previous model.fit calls
            stratify=y_train_for_models,
            random_state=seed # Use the current seed for reproducibility of val split
        )


        callbacks = [
            EarlyStopping(
                patience=5,
                restore_best_weights=True
            ),
            ReduceLROnPlateau(patience=3)
        ]

        model.fit(
            X_train_fold, y_train_fold,
            validation_data=(X_val_fold, y_val_fold),
            epochs=50,
            batch_size=32,
            verbose=0,
            callbacks=callbacks
        )

        auc, precision, recall, f1 = evaluate_model(model, X_test_data, y_test_for_models)

        all_results.append([
            model_name,
            seed,
            auc,
            precision,
            recall,
            f1
        ])
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

df_transformer = pd.read_csv("tcga_brca_sequence_dataset.csv")

token_map = {"A":0, "T":1, "G":2, "C":3}
def tokenize(seq):
    return [token_map[b] for b in seq]
X_tok_transformer = np.array([tokenize(s) for s in df_transformer["sequence"]])

df_transformer["label"] = df_transformer["type"].apply(lambda x: 0 if "Silent" in x else 1)
y_transformer = df_transformer["label"].values

def build_transformer(seq_len):
    # Re-build the transformer model structure
    from tensorflow.keras.layers import Input, GlobalAveragePooling1D, Dense, Dropout, Layer
    from tensorflow.keras.models import Model

    # Positional Embedding Layer from previous cells
    class PositionalEmbedding(Layer):
        def __init__(self, length, vocab_size, embed_dim):
            super().__init__()
            self.token_emb = tf.keras.layers.Embedding(vocab_size, embed_dim)
            self.pos_emb = tf.keras.layers.Embedding(length, embed_dim)

        def call(self, x):
            positions = tf.range(start=0, limit=tf.shape(x)[-1])
            return self.token_emb(x) + self.pos_emb(positions)

    # Transformer block from previous cells
    class TransformerBlock(Layer):
        def __init__(self, embed_dim, num_heads, ff_dim):
            super().__init__()
            self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
            self.ffn = tf.keras.Sequential([
                Dense(ff_dim, activation="relu"),
                Dense(embed_dim)
            ])
            self.norm1 = LayerNormalization()
            self.norm2 = LayerNormalization()
            self.drop1 = Dropout(0.1)
            self.drop2 = Dropout(0.1)

        def call(self, x):
            attn = self.att(x, x)
            x = self.norm1(x + self.drop1(attn))
            ffn = self.ffn(x)
            return self.norm2(x + self.drop2(ffn))


    SEQ_LEN = seq_len # Use the dynamic seq_len passed to the builder
    VOCAB = 4
    EMBED = 64

    inputs = Input(shape=(SEQ_LEN,))
    x = PositionalEmbedding(SEQ_LEN, VOCAB, EMBED)(inputs)

    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)

    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)

    outputs = Dense(1, activation="sigmoid")(x)

    transformer_model = Model(inputs, outputs)

    transformer_model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return transformer_model

auc_transformer = train_eval_auc(
    build_transformer,
    X_tok_transformer,
    y_transformer
)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score
from scipy import stats
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Input, GlobalAveragePooling1D, Dense, Dropout, Layer, LayerNormalization, MultiHeadAttention, Conv1D, MaxPooling1D, BatchNormalization, Activation, Add, Embedding
from tensorflow.keras.models import Sequential, Model

# --- Model Builder Functions (re-defined here for self-containment) ---

def build_cnn(seq_len):
    model = Sequential([
        Conv1D(64, 7, activation="relu", input_shape=(seq_len, 4)),
        MaxPooling1D(2),
        Conv1D(128, 5, activation="relu"),
        MaxPooling1D(2),
        Conv1D(256, 3, activation="relu"),
        MaxPooling1D(2),
        Flatten(),
        Dense(256, activation="relu"),
        Dropout(0.5),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

class PositionalEmbedding(Layer):
    def __init__(self, length, vocab_size, embed_dim):
        super().__init__()
        self.token_emb = Embedding(vocab_size, embed_dim)
        self.pos_emb = Embedding(length, embed_dim)
    def call(self, x):
        positions = tf.range(start=0, limit=tf.shape(x)[-1])
        return self.token_emb(x) + self.pos_emb(positions)

class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm1 = LayerNormalization()
        self.norm2 = LayerNormalization()
        self.drop1 = Dropout(0.1)
        self.drop2 = Dropout(0.1)
    def call(self, x):
        attn = self.att(x, x)
        x = self.norm1(x + self.drop1(attn))
        ffn = self.ffn(x)
        return self.norm2(x + self.drop2(ffn))

def build_transformer(seq_len):
    VOCAB = 4
    EMBED = 64
    inputs = Input(shape=(seq_len,))
    x = PositionalEmbedding(seq_len, VOCAB, EMBED)(inputs)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = TransformerBlock(EMBED, num_heads=4, ff_dim=128)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    transformer_model = Model(inputs, outputs)
    transformer_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return transformer_model

def cnn_feature_extractor(x_input):
    x1 = Conv1D(64, 7, padding="same")(x_input)
    x1 = BatchNormalization()(x1)
    x1 = Activation("relu")(x1)
    x1 = MaxPooling1D(2)(x1)
    x2 = Conv1D(128, 5, padding="same")(x1)
    x2 = BatchNormalization()(x2)
    x2 = Activation("relu")(x2)
    x2 = MaxPooling1D(2)(x2)
    x3 = Conv1D(256, 3, padding="same")(x2)
    x3 = BatchNormalization()(x3)
    x3 = Activation("relu")(x3)
    shortcut = Conv1D(256, 1, padding="same")(x2)
    x3 = Add()([x3, shortcut])
    x3 = Dropout(0.3)(x3)
    return x3

# PositionalEncoding for hybrid is different (for features, not tokens)
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, max_len, embed_dim, **kwargs):
        super().__init__(**kwargs)
        self.pos_emb = Embedding(input_dim=max_len, output_dim=embed_dim)
    def call(self, x):
        seq_len = tf.shape(x)[1]
        positions = tf.range(start=0, limit=seq_len, delta=1)
        pos_encoding = self.pos_emb(positions)
        return x + pos_encoding

class HybridTransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim):
        super().__init__()
        self.att = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim // num_heads, dropout=0.3)
        self.ffn = tf.keras.Sequential([Dense(ff_dim, activation="relu"), Dropout(0.3), Dense(embed_dim)])
        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.norm2 = LayerNormalization(epsilon=1e-6)
        self.drop1 = Dropout(0.3)
        self.drop2 = Dropout(0.3)
    def call(self, x, training=False):
        attn_output = self.att(x, x)
        attn_output = self.drop1(attn_output, training=training)
        x = self.norm1(x + attn_output)
        ffn_output = self.ffn(x, training=training)
        ffn_output = self.drop2(ffn_output, training=training)
        return self.norm2(x + ffn_output)

def build_hybrid(seq_len):
    inputs = Input(shape=(seq_len, 4))
    x = cnn_feature_extractor(inputs)
    x = PositionalEncoding(max_len=125, embed_dim=256)(x)
    x = HybridTransformerBlock(embed_dim=256, num_heads=2, ff_dim=128)(x)
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation="relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    outputs = Dense(1, activation="sigmoid")(x)
    hybrid = Model(inputs, outputs)
    hybrid.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), loss="binary_crossentropy", metrics=["accuracy"])
    return hybrid

def evaluate_model(model, X_test, y_test):
    y_pred_prob = model.predict(X_test).ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)

    auc = roc_auc_score(y_test, y_pred_prob)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    return auc, precision, recall, f1

# ----------------------------
# Reproducibility
# ----------------------------
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ----------------------------
# Hyperparameters
# ----------------------------
EPOCHS = 50
BATCH_SIZE = 32

# ----------------------------
# 10-Fold Stratified CV
# ----------------------------
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=SEED)

cnn_aucs = []
hybrid_aucs = []
transformer_aucs = []

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=0
)

print("Starting 10-Fold Cross Validation...\n")

for fold, (train_idx, val_idx) in enumerate(skf.split(X_balanced_oh_full, y_balanced_full_labels)): # Use _oh_full for splitting

    print(f"Fold {fold+1}/10")

    # Prepare data for CNN and Hybrid (one-hot encoded)
    X_train_oh, X_val_oh = X_balanced_oh_full[train_idx], X_balanced_oh_full[val_idx]
    # Prepare data for Transformer (tokenized)
    X_train_tok, X_val_tok = X_tok_balanced_full[train_idx], X_tok_balanced_full[val_idx]

    y_train, y_val = y_balanced_full_labels[train_idx], y_balanced_full_labels[val_idx]

    # ---------------- CNN ----------------
    cnn_model = build_cnn(X_balanced_oh_full.shape[1]) # Pass correct seq_len
    cnn_model.fit(
        X_train_oh, y_train,
        validation_data=(X_val_oh, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=0
    )
    y_pred_cnn = cnn_model.predict(X_val_oh).ravel() # Predict with correct data
    cnn_auc = roc_auc_score(y_val, y_pred_cnn)
    cnn_aucs.append(cnn_auc)

    # ---------------- Hybrid ----------------
    hybrid_model = build_hybrid(X_balanced_oh_full.shape[1]) # Pass correct seq_len
    hybrid_model.fit(
        X_train_oh, y_train,
        validation_data=(X_val_oh, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=0
    )
    y_pred_hybrid = hybrid_model.predict(X_val_oh).ravel() # Predict with correct data
    hybrid_auc = roc_auc_score(y_val, y_pred_hybrid)
    hybrid_aucs.append(hybrid_auc)

    # ---------------- Transformer ----------------
    transformer_model = build_transformer(X_tok_balanced_full.shape[1]) # Pass correct seq_len
    transformer_model.fit(
        X_train_tok, y_train,
        validation_data=(X_val_tok, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=0
    )
    y_pred_transformer = transformer_model.predict(X_val_tok).ravel() # Predict with correct data
    transformer_auc = roc_auc_score(y_val, y_pred_transformer)
    transformer_aucs.append(transformer_auc)

    print(f"  CNN AUC: {cnn_auc:.4f}")
    print(f"  Hybrid AUC: {hybrid_auc:.4f}")
    print(f"  Transformer AUC: {transformer_auc:.4f}")
    print("-" * 40)

# Convert to numpy
cnn_aucs = np.array(cnn_aucs)
hybrid_aucs = np.array(hybrid_aucs)
transformer_aucs = np.array(transformer_aucs)

# ----------------------------
# Summary Statistics
# ----------------------------
print("\n===== FINAL RESULTS (10-Fold CV) =====")
print(f"CNN: Mean={cnn_aucs.mean():.4f}, Std={cnn_aucs.std():.4f}")
print(f"Hybrid: Mean={hybrid_aucs.mean():.4f}, Std={hybrid_aucs.std():.4f}")
print(f"Transformer: Mean={transformer_aucs.mean():.4f}, Std={transformer_aucs.std():.4f}")

# ----------------------------
# Statistical Testing
# ----------------------------
def cohens_d(x, y):
    diff = x - y
    return np.mean(diff) / np.std(diff, ddof=1)

print("\n===== STATISTICAL TESTS =====")

# CNN vs Transformer
t_ct, p_ct = stats.ttest_rel(cnn_aucs, transformer_aucs)
w_ct, pw_ct = stats.wilcoxon(cnn_aucs, transformer_aucs)
d_ct = cohens_d(cnn_aucs, transformer_aucs)

print("\nCNN vs Transformer")
print(f"Paired t-test p-value: {p_ct:.6f}")
print(f"Wilcoxon p-value: {pw_ct:.6f}")
print(f"Cohen's d: {d_ct:.4f}")

# CNN vs Hybrid
t_ch, p_ch = stats.ttest_rel(cnn_aucs, hybrid_aucs)
w_ch, pw_ch = stats.wilcoxon(cnn_aucs, hybrid_aucs)
d_ch = cohens_d(cnn_aucs, hybrid_aucs)

print("\nCNN vs Hybrid")
print(f"Paired t-test p-value: {p_ch:.6f}")
print(f"Wilcoxon p-value: {pw_ch:.6f}")
print(f"Cohen's d: {d_ch:.4f}")
```
